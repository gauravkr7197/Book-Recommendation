```python
import matplotlib.pyplot as plt
```

```python
x=[500,1000,1500,2000,2500]
y=[13.985342400000036,28.43085910000002,50.4120572999999,73.65644359999999,99.47835360000005
]
```

```python
plt.title('strip headers') 
plt.xlabel('Batch size') 
# naming the y axis 
plt.ylabel('time taken')
plt.plot(x,y)
  
# function to show the plot 
plt.show()
```

```python
x1=[500,1000,1500,2000,2500]
y1=[0.043,0.122,0.201,0.400,0.619]
```

```python
plt.title('KNN') 
plt.xlabel('Batch size') 
# naming the y axis 
plt.ylabel('time taken for predictions')
plt.plot(x1,y1,color='green')
  
# function to show the plot 
plt.show()
```

```python
x2=[500,1000,1500,2000,2500]
y2=[05.122300,09.095703,15.768821,17.064627,23.275852]
plt.title('Preprocessing') 
plt.xlabel('Batch size') 
# naming the y axis 
plt.ylabel('time taken preprocessing')
plt.plot(x2,y2,color='red')
  
# function to show the plot 
plt.show()
```

```python
plt.plot(x,y,label="strip headers")
plt.plot(x1,y1,label="KNN")
plt.plot(x2,y2,label="preprocessing")
plt.legend()
plt.show()
```
